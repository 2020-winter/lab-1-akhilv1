```python
def countOnes(arr):
    c = 0;
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            if arr[i][j] == 1:
                c += 1
    return c,len(np.where(arr==1)[0])
countOnes(a), countOnes(b)
```
